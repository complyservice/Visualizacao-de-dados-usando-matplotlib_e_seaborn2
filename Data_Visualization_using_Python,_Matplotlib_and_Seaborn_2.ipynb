Visualização de dados

A visualização de dados é a representação gráfica dos dados. Envolve a produção de imagens que comunicam as relações entre os dados representados aos espectadores. A visualização de dados é uma parte essencial da análise de dados e do aprendizado de máquina, mas escolher o tipo certo de visualização geralmente é um desafio. Este guia fornece uma introdução às técnicas de visualização de dados populares, apresentando exemplos de casos de uso e fornecendo exemplos de código usando Python.

Tipos de gráficos cobertos:

    Gráfico de linha
    Gráfico de dispersão
    Histograma e Distribuição de Frequência
    Mapa de calor
    Gráfico de contorno
    Gráfico de caixa
    Gráfico de barras

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("darkgrid")
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Gráfico de linha

Um gráfico de linhas exibe informações como uma série de pontos de dados ou marcadores, conectados por linhas retas. Você pode personalizar a forma, tamanho, cor e outros elementos estéticos dos marcadores e linhas para melhor clareza visual.



Exemplo

Vamos criar um gráfico de linhas para comparar os rendimentos de maçãs e laranjas ao longo de 12 anos na região imaginária de Hoenn.

```python
#Dados de amostra
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896,]

#Primeira linha
plt.plot(years, apples, 'b-x', linewidth=4, markersize=12, markeredgewidth=4, markeredgecolor='navy')

#Segunda linha
plt.plot(years, oranges, 'r-o', linewidth=4, markersize=12,);

#Titulo
plt.title('Crop Yields in Hoenn Region')

#Rotulos de linha
plt.legend(['Apples', 'Oranges'])

#Axis Labels
plt.xlabel('Year') ; plt.ylabel('Yield(tons)');

```

Scatter Plot

Em um gráfico de dispersão, os valores de 2 variáveis ​​são plotados como pontos em uma grade bidimensional. Além disso, você também pode usar uma terceira variável para determinar o tamanho ou a cor dos pontos.

Exemplo
O conjunto de dados de flores Iris fornece medições de amostras de sépalas e pétalas para 3 espécies de flores. O conjunto de dados Iris está incluído na biblioteca seaborn e pode ser carregado como um dataframe pandas.


```python
# Carregar dados em um dataframe do Pandas
data = sns.load_dataset("iris")

#Ver os dados

data.sample(5)
```



Podemos usar um gráfico de dispersão para visualizar o comprimento da sépala e a largura da sépala variar em diferentes espécies de flores. Os pontos para cada espécie formam um cluster separado, com alguma sobreposição entre as espécies Versicolor e Virginica

```python
#Criando um Scatter Plot
sns.scatterplot(data.sepal_length, #X-axis
                data.sepal_width, #Y-axis
                hue=data.species, #Dot color
                s=100);
#Titulo do Grafico
plt.title("Flowers");
```

Histograma e Distribuição de Frequência

Um histograma representa a distribuição de dados formando bins ao longo do intervalo dos dados e, em seguida, desenhando barras para mostrar o número de observações que caem em cada bin.

Exemplo

Podemos usar um histograma para visualizar como os valores da largura da sépala são distribuídos.

```python
plt.title("Distribuition of Sepal Width")
sns.displot(data.sepal_width, kde=False);
```

Podemos ver imediatamente que os valores da largura da sépala estão na faixa de 2,0 a 4,5, e cerca de 35 valores estão na faixa de 2,9 a 3,1. Também podemos olhar para esses dados como uma distribuição de frequência, onde os valores no eixo Y são porcentagens em vez de contagens.

```python
plt.title("Distribuition of Sepal Width")
sns.displot(data.sepal_width);
```

Heatmap

Um mapa de calor
Um heatamp é usado para visualizar dados bidimensionais como uma matriz ou uma tabela usando cores.

Exemplo 
Usaremos outro conjunto de dados de amostra da Seaborn, chamado "voos", para visualizar o movimento mensal de passageiros em um aeroporto ao longo de 12 anos.

```python
# Carrega o conjunto de dados de voos de exemplo como uma matriz

flights = sns.load_dataset("flights").pivot("month","year", "passengers")
#Titulo do Grafico

plt.title("Number of Passengers(1000s)")
#Desenhe um mapa de calor com os valores numéricos em cada célula

sns.heatmap(flights,
            fmt="d",
            annot=True,
            linewidths=.5,
            cmap='Blues',
            annot_kws={"fontsize":13});
```

Gráfico de contorno

O gráfico de contorno usa contornos ou regiões codificadas por cores nos ajuda a visualizar 3 variáveis ​​numéricas em duas dimensões. Uma variável é representada no eixo horizontal e uma segunda variável é representada no eixo vertical. A terceira variável é representada por um gradiente de cores e isolinhas (linhas de valor constante).

Exemplo

Podemos visualizar os valores de largura e comprimento da sépala a partir do conjunto de dados de flores usando um gráfico de contorno. O tom de azul representa a densidade de valores em uma região do gráfico.

```python
plt.title("Flowers")
sns.kdeplot(data.sepal_length, data.sepal_width, shade=True, shade_lowest=False);
```

Podemos segmentar espécies de flores criando vários gráficos de contorno com cores diferentes.

```python
setosa = data[data.species == 'setosa']
virginica = data[data.species == 'virginica']

plt.title("Flowers (Setosa & Virginica)")

sns.kdeplot(setosa.sepal_length, setosa.sepal_width, shade=True, cmap='Reds', shade_lowest=False)
sns.kdeplot(virginica.sepal_length, virginica.sepal_width, shade=True, cmap='Blues', shade_lowest=False);
```

Gráfico de caixa(BOX PLOT)

Um gráfico de caixa mostra a distribuição de dados ao longo de um único eixo, usando uma "caixa" e "bigodes". A extremidade inferior da caixa representa o 1º quartil (ou seja, 25% dos valores estão abaixo dele), e a extremidade superior da caixa representa o 3º quartil (ou seja, 25% dos valores estão acima). O valor mediano é representado por uma linha dentro da caixa. Os "bigodes" representam os valores mínimo e máximo (às vezes excluindo valores discrepantes, que são representados como pontos).

Exemplo

Usaremos outro conjunto de dados de amostra incluído no Seaborn, chamado "dicas". O conjunto de dados contém informações sobre sexo, hora do dia, conta total e valor da gorjeta para clientes que visitam um restaurante durante uma semana.

```python
tips = sns.load_dataset("tips");
tips
```

Podemos usar um gráfico de caixa para visualizar a distribuição do total da conta para cada dia da semana, segmentado pelo fato de o cliente ser fumante.

```python
plt.title("Daily Total Bill")

# Desenhe um boxplot aninhado para mostrar as contas por dia e hora

sns.boxplot(tips.day, tips.total_bill, hue=tips.smoker);
```

Gráfico de barras (Bar Chart)

Um gráfico de barras apresenta dados categóricos com barras retangulares com alturas proporcionais aos valores que representam. Se houver vários valores para cada categoria, um gráfico de barras também pode representar o valor médio, com intervalos de confiança.

Exemplo

Podemos usar um gráfico de barras para visualizar o valor médio da fatura total para diferentes dias da semana, segmentados por sexo, para o conjunto de dados "dicas"

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips);
```
